```python
sns.pairplot(car,hue="Price")
```

```python
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
lr=LinearRegression()
ohe=OneHotEncoder()
```

```python
x=car.drop(columns='Price')
y=car['Price']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
ohe.fit(car[['name','company','fuel_type']])
ohe.categories_
```

```python
column_transf=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                      remainder='passthrough')
```

```python
pipe=make_pipeline(column_transf,lr)
pipe.fit(xtrain,ytrain)
```

```python
pipe
```

```python
ypred=pipe.predict(xtest)
ypred_real=np.exp(ypred)
ypred_real
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```

```python
r2score=[]
for i in range(2000):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=i)
    lr1=LinearRegression()
    pipe=make_pipeline(column_transf,lr1)
    pipe.fit(xtrain,ytrain)
    pred=pipe.predict(xtest)
    r2score.append(r2_score(ytest,pred))
    
```

```python
np.argmax(r2score)
```

```python
r2score[np.argmax(r2score)]
```

# Ensemble Learning : Voting(Wisdom of Crowd)

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import VotingRegressor
lr3=LinearRegression()
dt3=DecisionTreeRegressor()
svr=Pipeline([
    ('scaler',StandardScaler(with_mean=False)),
    ('svr',SVR())
])
voting=VotingRegressor([
    ('lr',lr3),
    ('dt',dt3)
])
pipe3=Pipeline([
    ('preprocess',column_transf),
    ('model',voting)
])
```

```python
pipe3
```

```python
pipe3.fit(xtrain,ytrain)
```

```python
scores=cross_val_score(pipe3,xtrain,ytrain,scoring='r2',cv=10)
print("Voting R2:",np.round(np.mean(scores),2))
```
